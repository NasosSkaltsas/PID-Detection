```python
from ultralytics import YOLO
import shutil, os
```

```python
# Train a tiny model quickly
model = YOLO('yolov8n.pt')
```

```python

model.train(data='../../data/model.yaml', epochs=150, imgsz=640, batch=12, project='runs', name='detection_model',  exist_ok=True  )
```

```python
# Export best weights 
best = f'runs/detection_model/weights/best.pt'
os.makedirs('../../models', exist_ok=True)
shutil.copy(best, '../../models/object_detection/bestdetection_model.pt')
```
